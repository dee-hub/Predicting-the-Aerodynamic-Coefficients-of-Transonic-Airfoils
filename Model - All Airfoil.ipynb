# C5A

```python
cols = ["ID", "Mach No", "Angle of Attack", "Lift Coefficient", "Drag Coefficient"]
all_data_complete = pd.read_csv('c5a.csv', sep=",", header=None, names=cols)
all_data_c5a = all_data_complete.drop(all_data_complete.index[[0]])
all_data_c5a = all_data_c5a.reset_index(drop=True)
```

```python
all_data_c5a
```

```python
a = list(all_data_c5a['Angle of Attack'].unique())
```

```python
a
```

```python
values_c5a = all_data_c5a["Angle of Attack"].value_counts()
```

```python
values_c5a
```

```python
a = all_data_c5a['Angle of Attack'].unique()
```

```python
import os
import cv2
import glob
images_c5a = []
inputPath = 'c5a-images'
    # loop over the indexes of the houses
count = 0
airfoils = []
for i in a:
    # find the four images for the house and sort the file paths,
        # ensuring the four are always in the *same order*
    basePath = os.path.sep.join([inputPath, "{}_*".format(i)])
    housePaths = sorted(list(glob.glob(basePath)))
     #       # load the input image, resize it to be 32 32, and then
      #      # update the list of input images
    for housePath in housePaths:
        count+=1
        image = cv2.imread(housePath)
        image = cv2.resize(image, (200, 133))
        images_c5a.append(image)
        print(str(count) + ',images appended')
```

```python
images_c5a
```

```python
all_data_c5a = all_data_c5a.drop(["ID"], axis=1)
```

```python
all_data_c5a.replace('', np.nan, inplace=True)
```

## Sorting Missing Data

```python
a_c5a = all_data_c5a[all_data_c5a['Lift Coefficient'].isnull()].index
```

```python
for i in a_c5a:
    print(i)
```

```python
for i in a_c5a:
    c = i - 1
    first_num = float(all_data_c5a["Lift Coefficient"].iloc[c])
    print('first_number', first_num)
    d = i + 1
    second_num = float(all_data_c5a["Lift Coefficient"].iloc[d])
    print('second_number', second_num)
    average = first_num - second_num
    average/=2
    major_num = first_num - average
    print('major_number', major_num)
    all_data_c5a.at[i,'Lift Coefficient']=major_num
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import concatenate
import numpy as np
import argparse
import locale
import os

trainAttrX, testAttrX, trainImagesX, testImagesX = train_test_split(features_joined, images_joined, test_size=0.20, random_state=42)
```

```python
trainY = trainAttrX["Lift Coefficient"]
testY = testAttrX["Lift Coefficient"]
```

```python
trainAttrX = trainAttrX[["Mach No", "Angle of Attack", "Reynolds number"]]
testAttrX = testAttrX[["Mach No", "Angle of Attack", "Reynolds number"]]
```

```python
from sklearn.preprocessing import MinMaxScaler
cs = MinMaxScaler()
trainAttrX = cs.fit_transform(trainAttrX)
testAttrX = cs.fit_transform(testAttrX)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Activation
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Input
from tensorflow.keras.models import Model
def create_mlp(dim, regress=False):
    # define our MLP network
    model = Sequential()
    model.add(Dense(8, input_dim=dim, activation="relu"))
    model.add(Dense(4, activation="relu"))
    # check to see if the regression node should be added
    if regress:
        model.add(Dense(1, activation="linear"))
    # return our model
    return model

def create_cnn(width, height, depth, filters=(16, 32, 64), regress=False):
    # initialize the input shape and channel dimension, assuming
    # TensorFlow/channels-last ordering
    inputShape = (height, width, depth)
    chanDim = -1
    # define the model input
    inputs = Input(shape=inputShape)
    # loop over the number of filters
    for (i, f) in enumerate(filters):
        # if this is the first CONV layer then set the input
        # appropriately
        if i == 0:
            x = inputs
        # CONV => RELU => BN => POOL
        x = Conv2D(f, (3, 3), padding="same")(x)
        x = Activation("relu")(x)
        x = BatchNormalization(axis=chanDim)(x)
        x = MaxPooling2D(pool_size=(2, 2))(x)
        
# flatten the volume, then FC => RELU => BN => DROPOUT
        x = Flatten()(x)
        x = Dense(16)(x)
        x = Activation("relu")(x)
        x = BatchNormalization(axis=chanDim)(x)
        x = Dropout(0.5)(x)
	# apply another FC layer, this one to match the number of nodes
	# coming out of the MLP
        x = Dense(4)(x)
        x = Activation("relu")(x)
	# check to see if the regression node should be added
        if regress:
            x = Dense(1, activation="linear")(x)
    # construct the CNN
        model = Model(inputs, x)
        # return the CNN
        return model
```

```python
mlp = create_mlp(trainAttrX.shape[1], regress=False)
cnn = create_cnn(200, 133, 3, regress=False)
# create the input to our final set of layers as the *output* of both
# the MLP and CNN
combinedInput = concatenate([mlp.output, cnn.output])
# our final FC layer head will have two dense layers, the final one
# being our regression head
x = Dense(4, activation="relu")(combinedInput)
x = Dense(1, activation="linear")(x)
# our final model will accept categorical/numerical data on the MLP
# input and images on the CNN input, outputting a single value (the
# predicted price of the house)
model = Model(inputs=[mlp.input, cnn.input], outputs=x)
```

```python
opt = Adam(lr=1e-3, decay=1e-3 / 200)
model.compile(loss="mean_absolute_percentage_error", optimizer=opt, metrics=['mse', 'mae', 'cosine_proximity'])
# train the model
print("[INFO] training model...")
history_iteration1 = model.fit(x=[trainAttrX, trainImagesX], y=trainY,
          validation_data=([testAttrX, testImagesX], testY),
          epochs=20, batch_size=8)
```
